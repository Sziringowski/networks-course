```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 5), sharex=False)


def t_c_s(N):
    return max(N * 15_000_000 / 30_000, 15_000_000 / 2_000)

def t_p_to_p(N, u):
    return max(15_000_000 / 30_000, 15_000_000 / 2_000, N * 15_000_000 / (30_000 + N * u))

range = [10,100,1000]
array = [t_c_s(n) for n in range]
ax.plot(range, array, label = f'client - serv arch', marker='+')


for u in [300, 700, 2_000]:
    array = [t_p_to_p(n, u) for n in range]
    ax.plot(range, array, label = f'p2p arch u = {u} kbps', marker='+')
        
ax.grid(True, linestyle="--", alpha=0.7)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('number of peers')
ax.set_ylabel('time (s)')
plt.legend()
plt.show()
```
